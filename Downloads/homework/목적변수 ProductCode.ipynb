```python
ds_data.productCode.head()키
ds_data["productCode"].replace(to_replace="EIQ", value=0, inplace = True)
ds_data["productCode"].replace(to_replace="EMQ", value=1, inplace = True)
ds_data["productCode"].replace(to_replace="MIQ", value=2, inplace = True)
ds_data["productCode"].replace(to_replace="MMQ", value=3, inplace = True)
```

```python
# EIQ, EMQ, MIQ, MMQ
```

# EMQ EIQ MIQ MMQ

```python
#get_dummies : 데이터의 문자형 컬럼에 대한 더미변수 생성
ds_data = pd.get_dummies(ds_data)
ds_data_dummy = pd.get_dummies(ds_data)
```

```python
#dummy 변수로부터 변수 역할 지정
ds_data_y = ds_data_dummy["productCode"]
ds_data_x = ds_data_dummy.drop("productCode", axis = 1, inplace = False)
#train_test_split(X: 설명변수 데이터, Y: 목표변수 데이터, test_size = test 데이터비율, random_state : 랜덤)
ds_data_train_x, ds_data_test_x, ds_data_train_y, ds_data_test_y =train_test_split(ds_data_x, ds_data_y, test_size = 0.30, random_state = 1234)

print("train data X size : {}".format(ds_data_train_x.shape))
print("train data Y size : {}".format(ds_data_train_y.shape))
print("test data X size : {}".format(ds_data_test_x.shape))
print("test data Y size : {}".format(ds_data_test_y.shape))
```

```python
#train 데이터셋 스케일 조정
ds_data_train_x_scaled = scale(ds_data_train_x, axis = 0)

# test 데이터셋 스케일 조정
ds_data_test_x_scaled = scale(ds_data_test_x, axis = 0)

#스케일이 변경된 X변수 확인
pd.DataFrame(ds_data_train_x_scaled, columns = ds_data_train_x.columns).head()
```
