```python
df_personal = ds_data[ds_data['customerType'] == 'PERSONAL']
df_business = ds_data[ds_data['customerType'] == 'BUSINESS']
```

# 로지스틱회귀분석

```python
#train_test_split(X: 설명변수 데이터, Y: 목표변수 데이터, test_size = test 데이터비율, random_state : 랜덤)
df_personal_train, df_personal_test =train_test_split(df_personal, test_size = 0.30, random_state = 1234)

print("train data size : {}".format(df_personal_train.shape))
print("test data size : {}".format(df_personal_test.shape))



#train_test_split(X: 설명변수 데이터, Y: 목표변수 데이터, test_size = test 데이터비율, random_state : 랜덤)
df_business_train, df_business_test =train_test_split(df_business, test_size = 0.30, random_state = 1234)

print("train data size : {}".format(df_business_train.shape))
print("test data size : {}".format(df_business_train.shape))
```

```python
#train 데이터에 상수 추가
df_personal_train = sm.add_constant(df_personal_train)
df_personal_train.head()
#로지스틱 회귀모델 및 회귀계수 검토
```

```python
##gender
df_personal_train["gender"].replace(to_replace="FEMALE", value=0, inplace = True)
df_personal_train["gender"].replace(to_replace="MALE", value=1, inplace = True)
## age
df_personal_train["age"].replace(to_replace="20대", value=2, inplace = True)
df_personal_train["age"].replace(to_replace="30대", value=3, inplace = True)
df_personal_train["age"].replace(to_replace="40대", value=4, inplace = True)
df_personal_train["age"].replace(to_replace="50대", value=5, inplace = True)
df_personal_train["age"].replace(to_replace="60대 이상", value=6, inplace = True)
## marital
df_personal_train["marital"].replace(to_replace="MARRIED", value=1, inplace = True)
df_personal_train["marital"].replace(to_replace="SINGLE", value=0, inplace = True)
## credit
df_personal_train["credit"].replace(to_replace="LOW", value=0, inplace = True)
df_personal_train["credit"].replace(to_replace="MIDDLE", value=1, inplace = True)
df_personal_train["credit"].replace(to_replace="HIGHT", value=2, inplace = True)
## residence
df_personal_train["residence"].replace(to_replace="HOUSE", value=0, inplace = True)
df_personal_train["residence"].replace(to_replace="APART", value=1, inplace = True)
df_personal_train["residence"].replace(to_replace="OFFICETEL", value=2, inplace = True)
df_personal_train["residence"].replace(to_replace="ETC.", value=3, inplace = True)
## productCode
df_personal_train["productCode"].replace(to_replace="EIQ", value=0, inplace = True)
df_personal_train["productCode"].replace(to_replace="EMQ", value=1, inplace = True)
df_personal_train["productCode"].replace(to_replace="MIQ", value=2, inplace = True)
df_personal_train["productCode"].replace(to_replace="MMQ", value=3, inplace = True)
df_personal_train["productCode"].replace(to_replace="HIQ", value=4, inplace = True)
df_personal_train["productCode"].replace(to_replace="HMQ", value=5, inplace = True)
## payment
df_personal_train["payment"].replace(to_replace="CMS", value=0, inplace = True)
df_personal_train["payment"].replace(to_replace="CARD", value=1, inplace = True)
## channel_new_y
df_personal_train["channel_new_y"].replace(to_replace="렌탈재계약", value=0, inplace = True)
df_personal_train["channel_new_y"].replace(to_replace="방문판매", value=1, inplace = True)
df_personal_train["channel_new_y"].replace(to_replace="오프라인", value=2, inplace = True)
df_personal_train["channel_new_y"].replace(to_replace="인터넷", value=3, inplace = True)
df_personal_train["channel_new_y"].replace(to_replace="홈쇼핑", value=4, inplace = True)
```

```python
df_personal_train.isnull().sum(axis = 0)
```

```python
#from_formula 함수를 이용하여 변수 역할 지정
log_model = sm.Logit.from_formula("""Termination ~ gender + age + marital + residence + channel_new_y + credit + quantity + maxDay + overDue + gift + productCode""", df_personal_train)
#적합
log_result = log_model.fit(method= 'bfgs')
#결과 출력
print(log_result.summary())
```

```python
# residence(T.ETC), maxDay(1), maxDay(6), productCode,
#로지스틱 모델로 test 데이터 예측
y_pred = log_result.predict(df_personal_test)
#0과 1의 값을 가진 class로 변환
y_pred_class = (y_pred  >0.5).astype(int)
#상위 5건 출력
y_pred_class.head()
```

```python
from sklearn import metrics
print("Accuracy:{0:.3f}".format(metrics.accuracy_score(df_personal_test["Termination"], y_pred_class)))
```

```python
#설명변수 중요도
ds_logistic_coef = pd.DataFrame({"Coef":log_result.params.values[1:]}, index = log_model.exog_names[1:])
ds_logistic_coef.plot.barh(y = "Coef")
```

```python
#from_formula 함수를 이용하여 변수 역할 지정
log_model = sm.Logit.from_formula("""Termination ~ C(gender) + C(age) + C(marital)  + C(residence) + C(channel_new_y) + C(credit) + quantity + C(maxDay) + C(overDue) + C(gift) + C(productCode) + C(payment) + 1""", df_personal_train)
#적합
log_result = log_model.fit(method= 'bfgs')
#결과 출력
print(log_result.summary())
```

```python
#get_dummies : 데이터의 문자형 컬럼에 대한 더미변수 생성
df_personal_dummy = pd.get_dummies(df_personal)
df_business_dummy = pd.get_dummies(df_business)
```

```python
#dummy 변수로부터 변수 역할 지정
df_personal_y = df_personal_dummy["Termination"]
df_personal_x = df_personal_dummy.drop("Termination", axis = 1, inplace = False)
#train_test_split(X: 설명변수 데이터, Y: 목표변수 데이터, test_size = test 데이터비율, random_state : 랜덤)
df_personal_train_x, df_personal_test_x, df_personal_train_y, df_personal_test_y =train_test_split(df_personal_x, df_personal_y, test_size = 0.30, random_state = 1234)

print("train data X size : {}".format(df_personal_train_x.shape))
print("train data Y size : {}".format(df_personal_train_y.shape))
print("test data X size : {}".format(df_personal_test_x.shape))
print("test data Y size : {}".format(df_personal_test_y.shape))


#dummy 변수로부터 변수 역할 지정
df_business_y = df_business_dummy["Termination"]
df_business_x = df_business_dummy.drop("Termination", axis = 1, inplace = False)
#train_test_split(X: 설명변수 데이터, Y: 목표변수 데이터, test_size = test 데이터비율, random_state : 랜덤)
df_business_train_x, df_business_test_x, df_business_train_y, df_business_test_y =train_test_split(df_business_x, df_business_y, test_size = 0.30, random_state = 1234)

print("train data X size : {}".format(df_business_train_x.shape))
print("train data Y size : {}".format(df_business_train_y.shape))
print("test data X size : {}".format(df_business_test_x.shape))
print("test data Y size : {}".format(df_business_test_y.shape))

```

```python
#train 데이터셋 스케일 조정
df_personal_train_x_scaled = scale(df_personal_train_x, axis = 0)
df_business_train_x_scaled = scale(df_business_train_x, axis = 0)

# test 데이터셋 스케일 조정
df_personal_test_x_scaled = scale(df_personal_test_x, axis = 0)
df_business_test_x_scaled = scale(df_business_test_x, axis = 0)

#스케일이 변경된 X변수 확인
pd.DataFrame(df_personal_train_x_scaled, columns = df_personal_train_x.columns).head()
pd.DataFrame(df_business_train_x_scaled, columns = df_business_train_x.columns).head()
```
