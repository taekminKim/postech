```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#데이터 불러오기
ds_hmeq = pd.read_csv("../../실습화일/상품구매.csv")
ds_hmeq
```

```python
#items 중 index 0번이 갖고 있으면 True, 아니면 False
enc = TransactionEncoder()
arr = enc.fit_transform(X= ds_hmeq)
#위의 함수는 데이터를 행렬의 형태로 반환 -> pandas의 DataFrame의 형태로 변환
ds_pre = pd.DataFrame(arr, columns = enc.columns_)
ds_pre
```

```python
#최소 지지도
min_support = 0.007

#지지도 계산
ds_freq = apriori(ds_pre, min_support = min_support, use_colnames = True)
ds_freq.round(3)
```

```python
#기준 = 신뢰도, 최소 신뢰도 = 0.5
ds_asso_rule = association_rules(ds_freq, metric = "confidence", min_threshold = 0.5)
ds_asso_rule.round(3)
```
